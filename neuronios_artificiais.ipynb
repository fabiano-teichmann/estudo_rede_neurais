### FUNO SOMA


```python
x = [0.8 , 0.2]
w = [0.5, 0.5]

soma = 0 
for i in range(0, len(x)):
    soma += x[i] * w[i]
          
print(f" Exemplo de funo de soma com resultado {soma}")

x = [-1, 7, 5]
w = [0.8, 0.1, 0]
soma = 0 
for i in x:
    i = x.index(i)
    soma += x[i] * w[i]
          
print(f" Exemplo de funo de soma com resultado {soma}")
```

### Step function (funo degrau)
Maior do que zero = 1 o neuronio vai ser ativado
Caso contrrio = 0 o neuronio no vai ser ativado




```python
import numpy as np
# exemplo percebtron numpy
x = np.array([1, 7, 5])
w =  np.array([0.8, 0.1, 0])
# DOT product escalar
def soma(x, w):
    return x.dot(w)
```

```python
def step_function(s):
    if s >= 1:
        return 1
    return 0 
step_function(soma(x, w))
```

```python
x = np.array([-1, 7, 5])
w =  np.array([0.8, 0.1, 0])
soma(x, w)
```

## Algoritmo ajuste de pesos



```python
def treinar(entradas, saidas, pesos, taxa_aprendizado):
    erro_total = 1
    while erro_total != 0:
        erro_total = 0
        for i in range(len(saidas)):
            saida_calculada = calc_saida(np.asarray(entradas[i]))
            erro = abs(saidas[i] - saida_calculada)
            erro_total += erro
            for j in range(len(pesos)):
                pesos[j] = pesos[j] + (taxa_aprendizado * entradas[i][j] * erro)
                print(f"Peso atualizado {pesos[j]}")
        print(f"Total de erros {str(erro_total)}")
        
```

```python
entradas = np.array([[0,0], [0,1], [1,0], [1,1], [2, 2]])
saidas = np.array([0, 0, 0, 1, 1])
pesos = np.array([0.0, 0.0])
taxa_aprendizado = 0.1

treinar(entradas, saidas, pesos, taxa_aprendizado)
```
