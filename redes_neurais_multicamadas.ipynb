```python
entradas = np.array([[0,0],
                    [0,1],
                    [1,0],
                    [1,1]])
```

```python
saidas = np.array([[0], [1], [1], [0]])
```

```python
pesos0 = np.array([[-0.424, -0.740, -0.961],
                   [0.358, -0.577, -0.469]])
```

```python
pesos1 = np.array([[-0.017],[-0.893], [0.148]])
```

```python
epocas = 100
for e in range(epocas):
    camada_entrada = entradas
    soma_sinapse0 = np.dot(camada_entrada, pesos0)
    camada_oculta = sigmoid(soma_sinapse0)
```

```python
camada_oculta
```

```python
soma_sinapse1 = np.dot(camada_oculta, pesos1)
```

```python
soma_sinapse1
```

```python
camada_saida = sigmoid(soma_sinapse1)
print(camada_saida)
```

```python
# Calculo erro com 51 % de acerto pois 0.49 é o percentual de erro
erro_camadaSaida = saidas - camada_saida
print(erro_camadaSaida)
media_absuluta = np.mean(np.abs(erro_camadaSaida))
print(media_absuluta)
```
